```python
tags=['div_stats_keeper','div_stats_keeper_adv','div_stats_shooting','div_stats_passing','div_stats_passing_types','div_stats_gca','div_stats_defense','div_stats_possession','div_stats_playing_time','div_stats_misc']
```

```python
tag_no=0
for link in list_ele[1:]:
    browser=webdriver.Chrome()
    browser.get(link.get_attribute('href'))
    time.sleep(10)
    browser.execute_script("window.scrollTo(0,"+ '15700' +")")
    
    content=browser.find_element(By.ID, tags[tag_no])


    rows=content.find_elements(By.TAG_NAME,'tr')
    data=[]
    for row in rows[2:]:
        cells=row.find_elements(By.TAG_NAME,'td')
        stats=[]
        for item in cells:
            stats.append(item.text)
        data.append(stats)

    cols=[]
    col_names=rows[1].find_elements(By.TAG_NAME,'th')
    for names in col_names:
        cols.append(names.text)

    df=pd.DataFrame(data,columns=cols[1:])
    print(df.head(2))
    print(df.tail(2))

    df.to_csv('LaLiga2023-24/'+tags[tag_no][4:]+'.csv',index=False)
    print('LaLiga2023-24/'+tags[tag_no][4:]+'.csv')
    
    tag_no+=1
    browser.close()
```
